```python
def variable_selection_continual(dataframe, method='kendalltau'):
    from scipy import stats
    valid_id = []
    selected_var_df = pd.DataFrame(columns=['concept_id', 'correlation', 'pvalue', 'num_records'])
    for id in dataframe.concept_id.unique():
        id_df = dataframe[dataframe["concept_id"]==id]
        id_df = id_df[id_df["first_abnormal_date"].notnull()]
        id_df["date"] = (id_df["concept_date"] - id_df["cohort_start_date"]).dt.days
        id_df = id_df[["date", "concept_value"]]
        id_df.drop_duplicates()
        if(len(id_df)<2):
            pass
        elif method == 'kendalltau':
            correlation, p_value = stats.kendalltau(id_df["date"].tolist(), id_df["concept_value"].tolist())
        elif method == 'spearmanr':
            correlation, p_value = stats.spearmanr(id_df["date"].tolist(), id_df["concept_value"].tolist())
        elif method == 'pearson':
            correlation, p_value = stats.pearsonr(id_df["date"].tolist(), id_df["concept_value"].tolist())
        if (p_value < 0.05):
            var = {}
            var['concept_id'] = id
            var['correlation'] = correlation
            var['pvalue'] = p_value
            var['num_records'] = len(id_df)
            selected_var_df = selected_var_df.append(var, ignore_index=True)
    return selected_var_df
```
